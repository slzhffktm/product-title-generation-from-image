```python
df_other = pd.read_csv('data/other_test.csv')
```

```python
decoder_data = set(df_decoder['file_names'].values)
other_data = set(df_other['file_names'].values)
```

```python
decoder_data = set([int(d.split('.')[0]) for d in decoder_data])
other_data = set([int(d.split('.')[0]) for d in other_data])
```

```python
import random
```

```python
res = list(decoder_data.intersection(other_data))
random.shuffle(res)
```
